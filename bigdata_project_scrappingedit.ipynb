```python

import requests
import bs4
import html2text

all_links = ['https://www.geeksforgeeks.org/']  # the start page
the_count = 0  # page number to scrape
count = 1
while len(all_links) < 100:
    print('####################### New page #################################', the_count)
    res = requests.get(all_links[the_count])  # res.text
    the_count += 1
    soup = bs4.BeautifulSoup(res.text, 'lxml')
    for link in soup.find_all('a', href=True):
        if str(link['href']).startswith('https://') and str(link['href']) not in all_links and count < 100:
            print('<<<<<<<<<<<<<<<<link number>>>>>>>>>>', count)
            all_links.append(link['href'])
            print(link['href'])
            count += 1
    NOF = 1
    for link in all_links:
        res2 = requests.get(link)  # res.text
        soup2 = bs4.BeautifulSoup(res2.text, 'lxml')
        output = html2text.html2text(soup2.text)
        with open(f"url{NOF}.txt", 'w', encoding="utf-8") as file:
            file.write(output)
            NOF += 1
```

```python
import pandas as pd
#"C:\\Users\\khaled\\Desktop\\invert.txt","r"

path = "C:\\Year 4\\Big Data\\Project\\output.txt"
method='r'
listUrl= filename_wordscount

def decodeFormatFile(path,method):
    data = open(path, method)
    dataWordFormat=[]
    uniqueWords=[]
    for detailsWord in data.readlines():
        detailsWord = " ".join(detailsWord.split())
        dataWord=detailsWord.split(" ")
        uniqueWords.append(dataWord[0])
        dataWord[1]= dataWord[1].split(';')
        dataWord[1].pop()

        for i in range(0,len(dataWord[1])):
            dataWord[1][i]=dataWord[1][i].split(':')

        dataWordFormat.append(dataWord)
    data.close()
    #show dataWordFormat
# [['khaled', [['file1', '2'], ['file2', '0'], ['file3', '2']]],
#  ['ahmed', [['file1', '1'], ['file2', '4'], ['file3', '1']]],
#  ['programer', [['file1', '1'], ['file2', '2'], ['file3', '4']]]]
    return set(uniqueWords), dataWordFormat

def convertWordDataIntoFrame(uniqueWords,ListOfEachDataWords,listUrl):

    wordsDataFrame = pd.DataFrame(0,columns=list(uniqueWords),index=listUrl.keys())
    for sampleWord in ListOfEachDataWords:
        for repetWordOfEachFile in sampleWord[1]:
            wordsDataFrame.loc[repetWordOfEachFile[0],sampleWord[0]]=int(repetWordOfEachFile[1])

    return wordsDataFrame

def computeTF(wordDataFrame, listUrl):
    for url,val in listUrl.items():
        wordDataFrame.loc[url]=wordDataFrame.loc[url]/float(val)
    return wordDataFrame

def computeIDF(ListOfEachDataWord,uniqueWords,listUrl):
# The log of the number of documents divided by the number of documents 
# that contain the word w. Inverse data frequency determines the weight 
# of rare words across all documents in the corpus.
    import math
    N = len(listUrl)

        #show ListOfEachDataWord
# [['khaled', [['file1', '2'], ['file2', '0'], ['file3', '2']]],
#  ['ahmed', [['file1', '1'], ['file2', '4'], ['file3', '1']]],
#  ['programer', [['file1', '1'], ['file2', '2'], ['file3', '4']]]]
    idfDict = dict.fromkeys(uniqueWords, 0)

    for dataWord in ListOfEachDataWord:
        idfDict[dataWord[0]] =len(dataWord[1])
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict

def computeTFIDF(wordDataFrameUseTF, dictionaryIDF):
    for word,idfWeight in dictionaryIDF.items():
       wordDataFrameUseTF.loc[:,word]=wordDataFrameUseTF.loc[:,word]*idfWeight
    return wordDataFrameUseTF


########################## main ###################################3
uniqueWords , ListOfEachDataWord= decodeFormatFile(path,method)
#Print Data of Words as Frame
wordDataFrame=convertWordDataIntoFrame(uniqueWords,ListOfEachDataWord,listUrl)
print(wordDataFrame)

#compute TF 
wordDataFrameUseTF=computeTF(wordDataFrame, listUrl)
print(wordDataFrameUseTF)

#compute IDf 
dictionaryIDF=computeIDF(ListOfEachDataWord,uniqueWords,listUrl)
print(dictionaryIDF)

#compute Tf and IDf 
wordDataFrameUseTFIDF=computeTFIDF(wordDataFrameUseTF, dictionaryIDF)
print(wordDataFrameUseTFIDF)


```
